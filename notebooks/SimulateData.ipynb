```python
import pandas as pd
import time
import os
from datetime import datetime

SOURCE_FILE = "data/Titanic-Dataset.csv"  
LANDING_ZONE = "data/raw_events/"

os.makedirs(LANDING_ZONE, exist_ok=True)

df = pd.read_csv(SOURCE_FILE)

batch_size = 10

for i in range(0, len(df), batch_size):
    chunk = df.iloc[i:i+batch_size].copy()

    chunk["event_time"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    chunk["source"] = "SIMULATED_DEVICE"

    file_name = f"event_{datetime.now().strftime('%Y%m%d_%H%M%S_%f')}.json"
    file_path = os.path.join(LANDING_ZONE, file_name)

    chunk.to_json(file_path, orient="records", lines=True)

    print(f"Sent batch {i//batch_size + 1} | rows: {len(chunk)}")

    time.sleep(3)
```
